```python
# 한 파일 내에서 여러 sheet를 효율적으로 읽기
xlsx = pd.ExcelFile("./sample_gmc_xlsx.xlsx")
df1 = pd.read_excel(xlsx, "Sheet1")
df2 = pd.read_excel(xlsx, "Sheet2")
xlsx.close()
```

```python
# or
with pd.ExcelFile("./sample_gmc_xlsx.xlsx") as xlsx:
    df1 = pd.read_excel(xlsx, "Sheet1")
    df2 = pd.read_excel(xlsx, "Sheet2")
```

```python
# or
## dict can take any type of object
## Note that if the same parsing parameters are used for all sheets,
## a list of sheet names can simply be passed to read_excel with no loss in performance.
print(xlsx.sheet_names)  # list
df_dict = pd.read_excel("sample_gmc_xlsx.xlsx", xlsx.sheet_names)

type(df_dict)
```
